# Handling Missing Data

### Filtering Out Missing Data

```python
from numpy import nan as NA

data = pd.Series([1, NA, 3.5, NA, 7])

data.dropna()
```

```python
data[data.notnull()]
```

```python
data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA], [NA, NA, NA], [NA, 6.5, 3.]])

cleaned = data.dropna()
```

```python
data
```

Passing how='all' will only drows rows that are all NA:

```python
data.dropna(how='all')
```

To drop columns in the same way, pass axis=1:

```python
data[4] = NA
data
```

```python
data.dropna(axis=1, how='all')
```

You can indicate a threshold using the thresh argument:

```python
data.dropna(thresh=2)
```

### Filling In Missing Data

```python
df = pd.DataFrame(np.random.randn(7, 3))

df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
```

```python
df.fillna(0)
```

Calling fillna with a dict, you can use a different fill value for each column:

```python
df.fillna({1: 0.5, 2: 0})
```

```python
df = pd.DataFrame(np.random.randn(7, 3))

df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='ffill', limit=2)
```

### Transforming Data Using a Function or Mapping

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami', 'corned beef', 'Bacon',
                             'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}

lowercased = data['food'].str.lower()

data['animal'] = lowercased.map(meat_to_animal)
data
```

We can also pass a function to do all of the work:

```python
data['food'].map(lambda x: meat_to_animal[x.lower()])
```

### Renaming Axis Indexes

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)), index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
data
```

Like a Series, the acis indeces have a map method.

```python
transform = lambda x: x[:4].upper()

data.index = data.index.map(transform)
data
```

```python
data.rename(index=str.title, columns=str.upper)
```

```python
data.rename(index={'OHIO': 'INDIANA'}, columns={'three': 'peekaboo'})
```

Should you wish to modify a dataset in-place, pass inplace=True:

```python
data.rename(index={'OHIO': 'INDIANA'}, inplace=True)
```

### Discretization and Binning

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]

bins = [18, 25, 35, 60, 100]

cats = pd.cut(ages, bins)
cats
```

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

A parenthesis means that the side is open, while the square bracket means it is closed. You can change which side is closed by passing right=False.

```python
pd.cut(ages, [18, 26, 36, 61, 100], right=False)
```

You can pass your own bin names by passing a list or array to the labels option:

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']

pd.cut(ages, bins, labels=group_names)
```

If you pass an integer number of bins to but instead of explicit bin edges, it will compute equal-length bins based on the minimum and maximum values in the data.

```python
data = np.random.rand(20)

pd.cut(data, 4, precision=2)
```

Similar to cut, you can pass your own quantiles:

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])
```

### Computing Indicator/Dummy Variables

```python
df = pd.DataFrame({'key1': ['b', 'b', 'a', 'c', 'a', 'b'],
                  'data': range(6)})

pd.get_dummies(df['key1'])
```

```python
dummies = pd.get_dummies(df['key1'], prefix='key')

df_with_dummy = df[['data']].join(dummies)
df_with_dummy
```

If a row in a DataFrame belongs to multiple categories, things are a bit more complicated:

```python
names = ['movie_id', 'title', 'genres']

movies = pd.read_table('datasets/movielens/movies.dat', sep='::', header=None, names=names)
movies[:5]
```

```python
all_genres = []

for x in movies.genres:
    all_genres.extend(x.split('|'))
    
genres = pd.unique(all_genres)
genres
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))

dummies = pd.DataFrame(zero_matrix, columns=genres)

gen = movies.genres[0]

gen.split('|')
```

```python
dummies.columns.get_indexer(gen.split('|'))
```

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
    
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.head()
```

A useful recipe for statistical applications is to combine get_dummies with a discretization function like cut:

```python
np.random.seed(12345)

values = np.random.rand(10)
values
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]

pd.get_dummies(pd.cut(values, bins)).head()
```

```python
import re

text = 'foo    bar\t baz   \nquz'

re.split('\s+', text)
```

When you call re.split('\s+', text), the regular expression is first compiled, and then its split method is called on the passed text. You can compile the regex yourself with re.compile, forming a reusable regex object.

```python
regex = re.compile('\s+')

regex.split(text)
```

If, instead, you wanted to get a list of all patterns matching the regex, you can use the finall method:

```python
regex.findall(text)
```

```python
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com"""

pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'

regex = re.compile(pattern, flags=re.IGNORECASE)

regex.findall(text)
```

```python
m = regex.search(text)
m
```

```python
text[m.start():m.end()]
```

regex.match returns None if the pattern does not occur at the start of the string:

```python
print(regex.match(text))
```

sub will return a new string with occurences of the pattern replaced by a new string:

```python
print(regex.sub('REDACTED', text))
```

```python
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'

regex = re.compile(pattern, flags=re.IGNORECASE)

regex.findall(text)
```
